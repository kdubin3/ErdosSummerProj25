```python
# Importing pandas to work with data
import pandas as pd

# Creating empty dataframes to fill with the information we care about
expenses_yearly = pd.DataFrame()
expenses_monthly = pd.DataFrame()

# Making the columns for years and months for both dataframes
years = range(2015,2025,1)
months = range(1,13,1)

expenses_yearly['year'] = years

expenses_monthly_y = []
expenses_monthly_m = []
for y in years:
    for m in months:
        expenses_monthly_y.append(y)
        expenses_monthly_m.append(m)

# 2025 only includes expenses through April, so we only include it for the monthly dataset
expenses_monthly_y += 4*[2025]
expenses_monthly_m += [1,2,3,4]

expenses_monthly['year'] = expenses_monthly_y
expenses_monthly['month'] = expenses_monthly_m

# Calculating total expenditures by year and month
total_yearly = []
total_monthly = []

for year in years:
    # Calculating for the year
    expenses = pd.read_csv("cta_expenditures/CTA_Expenditures_" + str(year) + ".csv")
    expenses['PAYMENT AMOUNT'] = expenses['PAYMENT AMOUNT'].astype(str)
    expenses['PAYMENT AMOUNT'] = expenses['PAYMENT AMOUNT'].str.replace(',', '')
    expenses['PAYMENT AMOUNT'] = expenses['PAYMENT AMOUNT'].str.replace('$', '')
    expenses['PAYMENT AMOUNT'] = pd.to_numeric(expenses['PAYMENT AMOUNT'])
    total_yearly.append(expenses['PAYMENT AMOUNT'].sum())

    # Calculating by month
    expenses['PAYMENT DATE'] = pd.to_datetime(expenses['PAYMENT DATE'])
    expenses['month'] = expenses['PAYMENT DATE'].dt.month

    for month in months:
        expenses_this_month = expenses[expenses['month'] == month]
        total_monthly.append(expenses_this_month['PAYMENT AMOUNT'].sum())
    
# For convenience, we deal with monthly 2025 expenses separately
expenses = pd.read_csv("cta_expenditures/CTA_Expenditures_2025.csv")
expenses['PAYMENT DATE'] = pd.to_datetime(expenses['PAYMENT DATE'])
expenses['month'] = expenses['PAYMENT DATE'].dt.month

for month in [1,2,3,4]:
    expenses_this_month = expenses[expenses['month'] == month]
    total_monthly.append(expenses_this_month['PAYMENT AMOUNT'].sum())

# Storing calculations
expenses_yearly['expenses'] = total_yearly
expenses_monthly['expenses'] = total_monthly

# Saving dataframes as csv files
expenses_yearly.to_csv('expenses_yearly.csv', index = False)
expenses_monthly.to_csv('expenses_monthly.csv', index = False)
```
